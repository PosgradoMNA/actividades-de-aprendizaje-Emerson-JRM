Value Counts

el conteo de valores es el mejor metodo para detemrinar cuantos valores tenemos de cada caracteristica

```python
df['drive-wheels'].value_counts()
```

```python
drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts
```

cambiaremos el nombre de drive wheels

```python
drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```
