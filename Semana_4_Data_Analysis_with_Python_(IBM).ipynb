# Data Acquisition

Agregaremos las librerias que estaremos ocupando

```python
import pandas as pd
```

se agrega el link o url a una variable, en este caso "link"

```python
link = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

se utiliza la propiedas de pandas para dar lectura a un csv, especificando que la inforamcion no tiene headers

```python
df= pd.read_csv(link, header = None)
```

para imprimir el data frame solo debes de colocar df

```python
df
```

si quieres visualizar las primeras filas de datos puedes ocupar el df.head(5), en este caso las primeras 5

```python
df.head(5)
```

si deseas visualizar las ultimas 5 seria el siguiente codigo df.tail(5)

```python
df.tail(5)
```

para poder agregar los encabezados mejor conocidos como headers se tendra que agregar los headers manuales de la siguiente manera, primero se definen los header en un arreglo como se muestra a continuacion:

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

despues se hace la asignacion de los encabezados, y lo visulizamos conlos primero 5 renglones

```python
df.columns = headers
df.head(5)
```
